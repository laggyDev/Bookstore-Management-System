```python
import pandas as pd

# # Load data from the CSV file
file_name = 'best-selling-books.csv'  # Update this to the path of your CSV file if it's not in the same directory

# Define classes:
class Book:
    def __init__(self, title, authors, original_language, first_published, sales_millions, genre):
        self.title = title
        self.authors = authors
        self.original_language = original_language
        self.first_published = first_published
        self.sales_millions = sales_millions
        self.genre = genre

    def __repr__(self):
        return f"'{self.title}' by {self.authors} ({self.original_language}, {self.first_published}) - {self.sales_millions}M copies - {self.genre}"

class Bookshelf:
    def __init__(self):
        self.books = []

    def load_from_csv(self, file_name):
        df = pd.read_csv(file_name)
        for _, row in df.iterrows():
            self.add_book(Book(row['Book'], row['Author(s)'], 
                               row['Original language'], 
                               row['First published'],
                               row['Approximate sales in millions'], 
                               row['Genre']))

    def add_book(self, book):
        self.books.append(book)

    def remove_book(self, title):
        self.books = [book for book in self.books if book.title != title]

    def find_book(self, title):
        for book in self.books:
            if book.title == title:
                return book
        return None

    def list_books(self):
        for book in self.books:
            print(book)

# Initialize a Bookshelf instance
my_bookshelf = Bookshelf()

# # Load data from the CSV file
# file_name = 'best-selling-books.csv'
my_bookshelf.load_from_csv(file_name)

# Example: Listing all books
my_bookshelf.list_books()
```
